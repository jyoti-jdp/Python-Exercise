```python
# Write a Python program to swap first and last digits of a number.
def swap(num):
    str_num = str( num )
    digits = []
    i = 0
    while i < len(str_num):
        digits.append(str_num[i])
        i=i+1
    digits[0],digits[len(str_num)-1] = digits[len(str_num)-1],digits[0]
    joined_digits = ''.join( digits )
    return int( joined_digits )
print(swap(478625))
```
