```python
# 20. Write a Python program that returns a string that is n (non-negative integer) copies of a given string.
def larger_str(text,n):
    result = ''
    for i in range(n):
        result = result +text
    return result
print(larger_str('Sky',6))
```

```python
# 22. Write a Python program to count the number 4 in a given list.
def count_value(list_1,n):
    list_1 = ' '.join(list_1)
    count = {}
    for i in range(len(list_1)):
        g = count.keys(n)
        if i==n and n in g:
            count[i] = count[i]+1
        else:
            count[i]=0
list_1 =[1,5,12,4,6,5,8,4,6,3,12,4]
n = 4
print(count_value(list_1,n))
```
